```python
# Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar estilo de gráficos
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# Cargar el dataset
df = pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')

# Vista general del dataset
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['type'].value_counts().plot(kind='bar', title='Cantidad de Películas vs Series', color=['skyblue', 'salmon'])
plt.xlabel('Tipo de Contenido')
plt.ylabel('Cantidad')
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='barh', title='Top 10 Países con más Contenido')
plt.xlabel('Cantidad de títulos')
plt.ylabel('País')
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot(kind='line', title='Títulos por Año de Lanzamiento')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.grid(True)
plt.show()
```

*Analisis:* En este gráfico observamos que el número de títulos en Netflix aumentó considerablemente a partir de 2015, con un pico en 2019, posiblemente impulsado por el auge del streaming y la expansión global de la plataforma.

Después de 2019, se nota una disminución en los lanzamientos, lo cual puede estar relacionado con la pandemia por COVID-19, que afectó la producción de contenido a nivel mundial.

Esta tendencia sugiere que Netflix tuvo un periodo de fuerte crecimiento antes de enfrentar restricciones externas que impactaron su catálogo.
